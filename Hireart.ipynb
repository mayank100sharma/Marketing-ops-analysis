```python
df['Year'] = pd.DatetimeIndex(df['Date of Contact']).year
df['Month'] = pd.DatetimeIndex(df['Date of Contact']).month
df['Day'] = pd.DatetimeIndex(df['Date of Contact']).day
```

```python
df.head()
```

```python
m1 = df.groupby(df['Year']).count()['Client Name']
m1.plot.bar()
```

```python
m2 = df.groupby(df['Month']).count()['Client Name']
m2.plot.bar()
```
